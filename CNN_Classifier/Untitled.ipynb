```python
import torch
import glob
from PIL import Image
import torchvision.transforms.functional as TF
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt

from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
```

```python
def showImage(image):
    plt.figure()
    image = image.numpy()
    image = image.transpose(1, 2, 0)
    image = np.squeeze(image, axis=2)
    plt.imshow(image)
    plt.show()
```

```python
def readData(train=True, mix=False):
    
    if (train):
        typ="training"
    else:
        typ="testing"
        
    if (mix):
        fold="mnist_png_mixtrain50"
    else:
        fold="mnist_png"
        
    
    images = torch.zeros(0, 1, 28, 28)
    #images = np.zeros((0,1,28,28))
    
    label = []
    
    
    for i in range(10):
        count = 0
        for file in glob.glob(f'./{fold}/{typ}/{i}/*'):
            image = Image.open(file)
            image = TF.to_tensor(image)  
            #image = np.expand_dims(image, axis=0)
            image.unsqueeze_(0)
            images = torch.cat((images, image), 0)
            #images = np.concatenate((images,image),0)
            
            label.append(i)
            count = count + 1
            
            if (count>100):
                break
        #print(images.shape)
        #print(len(label))
    
    #images = images.numpy()
    print(label[22])
    showImage(images[22])
    #print(images[13].shape)
    
    #label = np.asarray(label, dtype=np.float32)
    #label = np.asarray(label)
    
    #labels = np.zeros((label.size, label.max()+1))
    #labels[np.arange(label.size),label] = 1  
    ##labels = torch.from_numpy(label)
    
    #labels = label
    images = images.numpy()
    #labels = labels.numpy()
        
    return images, label
```

```python
Xtr, Ytr = readData()
Xts, Yts = readData(train=False)
```

```python
train_dataset = []
test_dataset = []
for i in range(Xtr.shape[0]):
   train_dataset.append([Xtr[i], Ytr[i]])
for i in range(Xtr.shape[0]):
   test_dataset.append([Xts[i], Yts[i]])
```

```python
train_loader = DataLoader(dataset=train_dataset, batch_size=2, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=2, shuffle=False)
```

```python
img, lbl = next(iter(train_loader))
```

```python
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=5, stride=1, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))
        self.layer2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=5, stride=1, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))
        self.drop_out = nn.Dropout()
        self.fc1 = nn.Linear(7 * 7 * 64, 1000)
        self.fc2 = nn.Linear(1000, 10)
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.reshape(out.size(0), -1)
        out = self.drop_out(out)
        out = self.fc1(out)
        out = self.fc2(out)
        return out
```

```python
num_epochs = 5
num_classes = 10
batch_size = 2
learning_rate = 0.001
MODEL_STORE_PATH = './Model'
```

```python
model = ConvNet()

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
imgs, lbls = next(iter(train_loader))
```

```python
lbls.shape
lbls[1]
```

```python
# Train the model
total_step = len(train_loader)
loss_list = []
acc_list = []
for epoch in range(num_epochs):
    for i, (imgs, lbls) in enumerate(train_loader):
        # Run the forward pass
        outputs = model(imgs)
        loss = criterion(outputs, lbls)
        loss_list.append(loss.item())

        # Backprop and perform Adam optimisation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Track the accuracy
        total = lbls.size(0)
        _, predicted = torch.max(outputs.data, 1)
        correct = (predicted == lbls).sum().item()
        acc_list.append(correct / total)

        if (i + 1) % batch_size == 0:
            print('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}, Accuracy: {:.2f}%'
                  .format(epoch + 1, num_epochs, i + 1, total_step, loss.item(),
                          (correct / total) * 100))
```

```python
# Test the model
model.eval()
with torch.no_grad():
    correct = 0
    total = 0
    for images, labels in test_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    print('Test Accuracy of the model on the 10000 test images: {} %'.format((correct / total) * 100))

# Save the model and plot
torch.save(model.state_dict(), MODEL_STORE_PATH + 'conv_net_model.ckpt')
```

```python
y = []
y2 = []

count = 0
for epoch in range(num_epochs):
    for i, (imgs, lbls) in enumerate(train_loader):
        if (i + 1) % batch_size == 0:
            y.append(loss_list[i])
            count = count + 1
            
for epoch in range(num_epochs):
    for i, (imgs, lbls) in enumerate(train_loader):
        if (i + 1) % batch_size == 0:
            y2.append(acc_list[i])

x = range(count)
print(len(y))
print(x)


  
# plotting the points  
plt.plot(x, y, label="loss") 

  
# Place a legend to the right of this smaller subplot.
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.) 

# naming the x axis 
plt.xlabel('epochs') 
# naming the y axis 
plt.ylabel('loss') 
  
# giving a title to my graph 
plt.title('Loss Visualization During training') 
  
# function to show the plot 
plt.show() 

plt.plot(x, y2, label="accuracy") 

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.) 

# naming the x axis 
plt.xlabel('epochs') 
# naming the y axis 
plt.ylabel('accuracy') 

plt.show()
```
